```python
import requests 
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
from data import google_maps
import time 
from pandas.io.json import json_normalize
```

```python
df = pd.read_csv('data/train.csv',nrows=200000)
df['distance'] = np.nan
df['duration'] = np.nan
```

```python
api_key = google_maps.secret()['api_key']
```

```python
def get_distance(params):
    google_maps_url = 'https://maps.googleapis.com/maps/api/directions/json?origin=_OLA_,_OLI_&destination=_DLA_,_DLI_&key=_K_'
    google_maps_url = google_maps_url.replace('_OLA_', params['pickup_lat']).replace('_OLI_',params['pickup_long']).replace('_DLA_',params['dropoff_lat']).replace('_DLI_',params['dropoff_long']).replace('_K_',api_key)
    r = requests.get(google_maps_url)
    result = r.json()
    raw_results = {'distance':result['routes'][0]['legs'][0]['distance'],'duration':result['routes'][0]['legs'][0]['duration']}
    return raw_results
```

```python
for i in df.iloc[50000:149975].index:
    param= {'pickup_lat':str(df['pickup_latitude'].iloc[i]),
            'pickup_long':str(df['pickup_longitude'].iloc[i]),
            'dropoff_lat':str(df['dropoff_latitude'].iloc[i]),
            'dropoff_long':str(df['dropoff_longitude'].iloc[i])}
    try: 
        location = get_distance(param)
        df['distance'].iloc[i]= location['distance']['value']
        df['duration'].iloc[i]=location['duration']['value']
    except:
        pass
    time.sleep(.2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_df[['distance','duration', 'passenger_count']], train_df['fare_amount'], test_size=0.33, random_state=42)
```

```python
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
