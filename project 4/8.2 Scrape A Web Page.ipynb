The result should be data and output in an organized format. Each of the data entries should be in its own dictionary with matching keys.


```python
bs = BeautifulSoup(requests.get('http://www.enoughproject.org/take_action').text, 'lxml')
```

```python
soup_dict = {}
for a in ['h1','h2','p', 'a', 'head','title', 'body','b']:
    if (bs.findAll(a)):
        soup_dict[a]=bs.findAll(a)
```

```python
driver = webdriver.Chrome()
try: 
    driver.get('https://www.bellevue.edu/degrees/master/data-science-ms/')
    driver.maximize_window()
    curriculum = driver.find_element_by_xpath('//*[@id="site-wrapper"]/div[2]/nav[1]/ul/li[2]/a')
    driver.execute_script("return arguments[0].scrollIntoView();", curriculum)
    time.sleep(1)
    curriculum.click()
    time.sleep(1)
    class_names = []
    course_expander = driver.find_element_by_xpath('//*[@id="degree-curriculum-section"]/div[1]/section/ul')
    for course in course_expander.find_elements_by_tag_name('li'):
        class_names.append(course.text)
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="site-wrapper"]/div[2]/nav[1]/ul/li[1]/a').click()
    time.sleep(1)        
except:
    pass
driver.close()
```

```python
for course in class_names:
    print('I scraped the course: ' + course.strip())
```
