```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('data/train.csv', nrows=2000000)
```

```python
len(train)
```

```python
train.columns
```

```python
train=train.reset_index()
```

```python
aggregate = train.groupby(['srch_destination_id', 'hotel_cluster']).sum().reset_index()
```

```python
train = train[train.is_booking==1]
```

```python
#imputation of missing origin distance
train.orig_destination_distance.fillna(train.orig_destination_distance.mean(), inplace=True)
```

```python
import seaborn as sns
sns.heatmap(train.corr())
```

```python
train.hotel_cluster.unique()
```

```python
train = train.rename(columns={'index':'id'})
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop(['date_time','hotel_cluster','is_booking','srch_ci','srch_co', 'cnt'], axis=1), train.hotel_cluster, test_size=0.33, random_state=42)
```

```python
mod = RandomForestRegressor().fit(X_train, y_train)
```

```python
mod.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
```

```python
dtree = DecisionTreeRegressor().fit(X_train,y_train)
dtree.score(X_test, y_test)
```

```python
dtree.predict(X_test)
```

```python
test = pd.read_csv('data/test.csv')
```

```python
test_dropped_cols = test.drop(['date_time','srch_ci','srch_co'], axis=1)
```

```python
test_dropped_cols.orig_destination_distance.fillna(test_dropped_cols.orig_destination_distance.mean(), inplace=True)
```

```python
dtree_pred = dtree.predict(test_dropped_cols)
random_forest_pred = mod.predict(test_dropped_cols)
```

```python
dtree_pred = pd.DataFrame(dtree_pred)
random_forest_pred = pd.DataFrame(random_forest_pred)
```

```python
dtree_pred[0] = dtree_pred[0].astype('int')
random_forest_pred[0] = random_forest_pred[0].astype('int')
```

```python
dtree_pred = dtree_pred.rename(columns={0:'dtree_pred'})
random_forest_pred = random_forest_pred.rename(columns={0:'forest_pred'})
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg.score(X_test, y_test)
```

```python
sub = test_dropped_cols.merge(dtree_pred, left_index=True, right_index=True)
```

```python
sub = sub[['id','dtree_pred']]
```

```python
logreg_pred = logreg.predict(test_dropped_cols)
logreg_pred = pd.DataFrame(logreg_pred)
logreg_pred[0] = logreg_pred[0].astype('int')
logreg_pred = logreg_pred.rename(columns={0:'logred_pred'})
sub = sub.merge(logreg_pred, left_index=True, right_index=True)
```

```python
sub = sub.merge(random_forest_pred, left_index=True, right_index=True)
```

```python
sub['hotel_cluster'] = sub['dtree_pred'].astype('str') + ' ' + sub['forest_pred'].astype('str')
```

```python
sub['hotel_cluster'] = sub['hotel_cluster'] + ' ' + sub['logred_pred'].astype('str')
```

```python
final_sub = sub[['id', 'hotel_cluster']]
```

```python
final_sub.to_csv('data/submission.csv', index=False)
```
