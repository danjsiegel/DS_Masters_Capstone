```python
df = pd.read_csv('630 R\\dodgers.csv')
```

```python
df.head()
```

```python
df.cap = pd.get_dummies(df.cap).ix[:,0] #No= 0, Yes = 1
df.shirt = pd.get_dummies(df.shirt).ix[:,0]#No= 0, Yes = 1
df.fireworks = pd.get_dummies(df.fireworks).ix[:,0]#No= 0, Yes = 1
df.bobblehead = pd.get_dummies(df.bobblehead).ix[:,0]#No= 0, Yes = 1
df.skies = pd.get_dummies(df.skies).ix[:,0] #clear = 0, cloudy = 1
df.day_night = pd.get_dummies(df.day_night).ix[:,0] #Day = 0, night = 1
```

```python
df.head()
#Looking at only nights 
df = df[df['day_night']==1]
```

```python
import statsmodels.formula.api as smf
```

```python
df.columns
```

```python
def feat_targ(x):
    if (x<50000):
        return 0
    else:
        return 1
```

```python
df['target'] = df['attend'].apply(lambda x: feat_targ(x))
```

```python
total_res = smf.ols('target ~ opponent + month + day_of_week + skies', data=df).fit()
total_res.summary()
```
